```python
import pandas as pd
from pathlib import Path
import plotly.express as px
```

```python
path = "/Users/coes3/phd-labs/fraganal/data_source/ooni/"
```

```python
def data_prep(path):
    files = Path(path).glob('*.csv')
    dfs = []

    for i in files:
        data = pd.read_csv(i, header=0)
        data['cc'] = i.name.split('_')[-1].strip('.csv')
        dfs.append(data)

    df = pd.concat(dfs, ignore_index=True)
    df.rename(columns={"measurement_start_day":"date", "confirmed_count":"blocked_count", "ok_count":"success_count"}, inplace = True)

    columns = ['failure_count', 'measurement_count', 'success_count']
    df.drop(columns, inplace=True, axis=1)

    df = df[['date', 'category_code', 'blocked_count', 'anomaly_count', 'cc']]
    
    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
    
    #  filter for only 2022 data
    df = df.loc[(df['date'] >= '2022-01-01') & (df['date'] <='2022-12-31')]
    
    #  filter out 0 counts
    df = df.loc[df['blocked_count'] != 0]
    return df
```

```python
df = data_prep(path)
```

```python
def plot_graph(df, melt=False):
    
    if melt:
        df2 = pd.melt(df, id_vars = ['date', 'cc', 'category_code'], var_name='result_type', value_name='count')
        
        fig = px.bar(df2, x="date", y="count", color="result_type", facet_col="cc", facet_col_wrap=2, hover_data=['count'],barmode='group', template='simple_white')
        fig.update_xaxes(dtick="M1", tickformat="%b\n%Y")
        fig.show()
    else:
        fig = px.pie(df, x="date", y="blocked_count", color="category_code", facet_col="cc", facet_col_wrap=2, hover_data=['blocked_count'], template='simple_white')
        fig.update_xaxes(dtick="M1", tickformat="%b\n%Y")
        fig.show()
```

```python
plot_graph(df, True)
```
