```python
import pandas as pd
from pathlib import Path
import plotly.express as px

import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
path = "/Users/coes3/phd-labs/fraganal/data_source/radar/"
```

```python
def data_prep(proto_type, path):
    #  function to prep radar data in pandas
    
    files = Path(path).glob(f'{proto_type}Versions*.csv')
    dfs = []
    
    for i in files:
        data = pd.read_csv(i, index_col=None, header=0)
        data['cc'] = i.name.split('_')[-1].strip('.csv')
        dfs.append(data)
        
    df = pd.concat(dfs, ignore_index=True)
    
    if proto_type == "IP":
        df.rename(columns={"Serie_0 timestamps":"date", "Serie_0  I Pv4":"ipv4", "Serie_0  I Pv6":"ipv6"}, inplace = True)
        
    else:
        df.rename(columns={"Serie_0 timestamps":"date", "Serie_0  T L S 1.3":"tlsv1_3", "Serie_0  T L S 1.2":"tlsv1_2", 
                           "Serie_0  T L S  Q U I C":"tlsvquic", "Serie_0  T L S 1.0":"tlsv1_0", "Serie_0  T L S 1.1":"tlsv1_1"}, inplace = True)
    
    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
    df['date'] = df['date'].dt.tz_convert(None)
    
    #  filter for only 2022 data
    df = df.loc[(df['date'] >= '2022-01-01') & (df['date'] <='2022-12-31')]
    
    return df
```

```python
df_tls = data_prep('TLS', path)
```

```python
df_ip = data_prep('IP', path)
```

```python
def plot_graph(df):
    
    df2 = pd.melt(df, id_vars = ['date', 'cc'], var_name='version', 
              value_name='pct')
    
    #  filter out 0 counts
    df2 = df2.loc[df2['pct'] != 0]
    
    fig = px.line(df2, x="date", y="pct", color="version", facet_col="cc", facet_col_wrap=3, hover_data=['pct'], template='simple_white')
    fig.update_xaxes(dtick="M1", tickformat="%b\n%Y")
    fig.show()
```

```python
plot_graph(df_ip)
```

```python
plot_graph(df_tls)
```

# Section on Generate pie chart pairing of IP and TLS and adoption pct

```python
def plot_pie(df_list, cc):
    num = 0
    
    fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
    
    map_mv_status_colors ={'ipv4':'#1f77b4',
                           'ipv6':'#ff7f0e',
                           'tlsv1_0':'#8c564b',
                           'tlsv1_1':'#d62728',
                           'tlsv1_2':'#9467bd',
                           'tlsv1_3':'#2ca02c',
                           'tlsvquic':'#e377c2'
                          }
    
    for df in df_list:
        num += 1
        tmp = df.groupby('cc', as_index=False).mean().round(0)
        df2 = pd.melt(tmp, id_vars = 'cc', var_name='version', 
              value_name='pct')
    
        #  filter out 0 counts
        df2 = df2.loc[df2['pct'] != 0]
        df2 = df2.loc[df2['cc'] == cc]
        
        if 'ipv4' in df2['version'].values:
            fig.add_trace(go.Pie(labels=df2["version"], values=df2["pct"], name="IP Versions"),
                      1, num)
        else:
            fig.add_trace(go.Pie(labels=df2["version"], values=df2["pct"], name="TLS Versions"),
                      1, num)
    # Use `hole` to create a donut-like pie chart
    fig.update_traces(hole=.3, hoverinfo="label+percent+name")
    
    fig.for_each_trace(
        lambda t: t.update(
            hole=0.3,
            hoverinfo="label+percent+name",
            marker_colors=pd.Series(t.labels).map(map_mv_status_colors),
        ).values
        )

    fig.update_layout(
        title_text=f"Protocol Version Distribution 2022, {cc}",
        # Add annotations in the center of the donut pies.
        annotations=[dict(text='IP', x=0.21, y=0.5, font_size=18, showarrow=False),
                     dict(text='TLS', x=0.79, y=0.5, font_size=18, showarrow=False)])
                          
    fig.show()
```

```python
cc_list = list(df_ip['cc'].unique())
```

```python
for i in cc_list:
    plot_pie([df_ip, df_tls], i)
```

# Section on Generate avg total percent of protocol adoption(combining IPv6, TLS 1.3 and TLS Quic)

```python
def total_protocol_pct(df_ip, df_tls, cc):
    v6_pct = df_ip['ipv6'].loc[df_ip['cc'] == cc].mean()
    tls13_pct = df_tls['tlsv1_3'].loc[df_tls['cc'] == cc].mean()
    tlsq_pct = df_tls['tlsvquic'].loc[df_tls['cc'] == cc].mean()
    proto_total = round((tlsq_pct+tls13_pct+v6_pct)/2, 2)
    print(f'{cc}: {proto_total}')
    return proto_total
```

```python
for i in cc_list:
    total_protocol_pct(df_ip, df_tls, i)
```
